> # **(ii)	Is the data equally distributed according to the sex?**

> ## **Countplot Sex**

```python
### We are using countplot to answer this question 
plt.figure(figsize = (10,5))
sns.countplot(df['sex']) # Counting the total of each class in sex column
```

```python
df.sex.value_counts() # Uniue value count in colum sex
```

# **(iii)	What is the relationship between the bmi and charges?**

> ## **BMI Vs Charges Scatter Plot**

```python
### We are using scatter plot between bmi and charges to answer this question
plt.figure(figsize = (10,5))
sns.scatterplot(df["bmi"], df["charges"])
```

> ## **BMI Vs Charges Scatter Plot (Improved)**

```python
plt.figure(figsize = (10,5))
smokers = df["smoker"].unique()
colors = ["Reds", "Greens"]
for i, smoker in enumerate(smokers):
    temp = df[df["smoker"] == smoker]
    sns.scatterplot(temp["bmi"], temp["charges"], cmap = colors[i])
plt.legend(smokers)
```

> # **(iv)	The people of which region pay the maximum charges?**

> ## **Boxplot of Region Vs Charges**

```python
### We are using boxplot to answer this question so we can see the median charges and outliers if there are any

plt.figure(figsize = (10,5))
sns.boxplot(x = "region", y = "charges", data = df)
```
